```python
# ================================================
# 1. Imports
# ================================================
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Ridge, ElasticNet
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
```

```python
# ================================================
# 2. Load datasets (train + eval)
# ================================================
train_df = pd.read_csv("C:\\Users\\soaha\\Desktop\\ML-Projects\\House_price_prediction_endToEnd\\data\\processed\\feature_engineered_train.csv")
eval_df = pd.read_csv("C:\\Users\\soaha\\Desktop\\ML-Projects\\House_price_prediction_endToEnd\\data\\processed\\feature_engineered_eval.csv")
```

```python
# ================================================
# 3. Define target & features
# ================================================
target = "price"
cols_to_drop = [target, "median_list_price", "median_ppsf", "median_list_ppsf"]
X_train = train_df.drop(columns=cols_to_drop, errors='ignore')
y_train = train_df[target]

X_eval = eval_df.drop(columns=cols_to_drop, errors='ignore')
y_eval = eval_df[target]
print(f"Dropped: {[c for c in cols_to_drop if c != target]}")
```

```python
# ================================================
# 4. Standardization (fit on train, transform eval)
# ================================================
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_eval_scaled  = scaler.transform(X_eval)
```

```python
# ================================================
# 5. Train & Evaluate Models
# ================================================

# --- Linear Regression ---
lr = LinearRegression()
lr.fit(X_train_scaled, y_train)
y_pred_lr = lr.predict(X_eval_scaled)

print("Linear Regression:")
print(" MAE:", mean_absolute_error(y_eval, y_pred_lr))
print(" RMSE:", np.sqrt(mean_squared_error(y_eval, y_pred_lr)))
print(" R²:", r2_score(y_eval, y_pred_lr))
```

```python
# --- Ridge Regression ---
ridge = Ridge(alpha=1.0)
ridge.fit(X_train_scaled, y_train)
y_pred_ridge = ridge.predict(X_eval_scaled)

print("\nRidge Regression:")
print(" MAE:", mean_absolute_error(y_eval, y_pred_ridge))
print(" RMSE:", np.sqrt(mean_squared_error(y_eval, y_pred_ridge)))
print(" R²:", r2_score(y_eval, y_pred_ridge))
```

```python
# --- ElasticNet ---
elastic = ElasticNet(alpha=0.1, l1_ratio=0.5)
elastic.fit(X_train_scaled, y_train)
y_pred_elastic = elastic.predict(X_eval_scaled)

print("\nElasticNet Regression:")
print(" MAE:", mean_absolute_error(y_eval, y_pred_elastic))
print(" RMSE:", np.sqrt(mean_squared_error(y_eval, y_pred_elastic)))
print(" R²:", r2_score(y_eval, y_pred_elastic))
```
